```python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)

b0, b1, b2, b3, b4, b5 = symbols('b0, b1, b2, b3, b4, b5')
eqv1 = -4**5 * b5 + (-4)**4 * b4 + -4**3 * b3 + (-4)**2 * b2 + -4*b1 +b0 + 4268
eqv2 = (-3)**5 * b5 + 3**4 * b4 -3**3 * b3 +3**2 * b2 -3*b1 + b0 + 1227
eqv3 = -b5 + b4 - b3 + b2 -b1 + b0 + 17
eqv4 = b5 + b4 + b3 + b2 + b1 + b0 -17
eqv5 = 3**5 * b5 + 3**4 * b4 + 3**3 * b3 + 9*b2 + 3 * b1 + b0 - 1227
eqv6 = 4**5 * b5 + 4**4 * b4 + 64 * b3 + 16 * b2 + 4 * b1 + b0 - 4268
nonlinsolve([eqv1,eqv2,eqv3,eqv4,eqv5,eqv6],[b5,b4,b3,b2,b1,b0])
x = Symbol('x')
f1 = 3 * x ** 5 + 19 * x **3 -5 * x
plot((f1), xlim = [-1,1], ylim = [-8,8])
solve(f1)
#Нечетная, - от -беск до 4 нуля, от 1 нуля до 5 нуля
```

```python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = Symbol('x')
```

## 1.
### Получите значение MSE меньшее 5.

```python
def print_points_ands_function1(sympy_function):
    def function(x_): return float(sympy_function.subs(x, x_))
    
    points_X = np.array([-2, -1, 0, 1, 2, 3, 3.5, 4, 4.5, 5])
    points_Y = np.array([2, -4, 1, 8, 21, 40, 47, 65, 75, 92])
    plt.xlim(-6, 10)
    plt.ylim(-1, 100)
    
    plt.scatter(points_X, points_Y, c='r')
    x_range = np.linspace(plt.xlim()[0], plt.xlim()[1], num=100)
    function_Y = [function(x_) for x_ in x_range]
    plt.plot(x_range, function_Y, 'b')
    plt.show()
    
    MSE = sum([(points_Y[i] - function(points_X[i]))**2 for i in range(len(points_Y))]) / len(points_Y)
    print(f'MSE = {MSE}')
```

```python
f1 = 48 * x**2 - 128 * x + 83
f1
f2 = expand(f1.subs(x,x + 1.5))
f2
f3 = expand(f2.subs(x,x/4))
f3
```

```python
print_points_ands_function1(f3)
```
