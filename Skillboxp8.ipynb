```python
a2 = Symbol('a2')
a1 = Symbol('a1')
a0 = Symbol('a0')
MSE = 1/3 * (((560 * a2 + 1945 * a1 + a0) - 4.3)**2 + ((340 * a2 + 1495 * a1 + a0) - 4.9)**2 + ((343 * a2 + 1570 * a1 + a0) - 5.2) ** 2)
```

```python
f2 = diff(MSE,a2)
f2
```

```python
f1 = diff(MSE,a1)
f1
```

```python
f0 = diff(MSE, a0)
f0
```

```python
MSEf2 = f2/3
MSEf2
```

```python
MSEf1 = f1/3
MSEf1
```

```python
MSEf0 = f0/3
MSEf0
```
